```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
ctr=0
linkd=pd.read_csv("D:/IndicWiki/120001-end.csv")
cols = ['CIN','Company Name','Company Status','RoC','Registration Number','Company Category','Company Sub Category','Class of Company','Date of Incorporation','Age of Company','Activity']
data_frame = pd.DataFrame(columns=cols)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
for i in range(1000):
    link=linkd.at[i,"url"]
    driver.get(link)
    try:
        table= driver.find_elements(by = By.XPATH,value="//table[@class='table table-striped']/tbody/tr")
        d={}
        d["CIN"]=(driver.find_elements(by = By.XPATH,value="//table[@class='table table-striped']/thead/tr/td/p/a"))[0].text
        for i in range(10):
            val=table[i].text.split("\n")
            if len(val)>1:
                v1=val[0]
                v2=val[1]
                if v1.startswith("Login") or v2.startswith("Login"):
                    continue
                d[v1]=v2
        data_frame=data_frame.append(d,ignore_index=True)
        ctr+=1
    except:
        pass
print(ctr)
data_frame
```

```python
data_frame.to_csv("D:/IndicWiki/Final_companies_result.csv")
```
